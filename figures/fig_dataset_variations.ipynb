```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load wandb_go.csv
df = pd.read_csv('/home/antoine/gene_pheno_pred/figures/wandb_base.csv')

# Assuming you have a DataFrame called 'df' with columns 'Hit@1' and 'method'

# Set up the plot
plt.figure(figsize=(6, 4))

# Create the swarmplot
method_order = ['TransE', 'TorusE', 'ComplEx', 'ConvKB']
sns.swarmplot(x='method', y='Hit@1', data=df, order=method_order)
# Add a horizontal line at 0.1 with the text 'Baseline'
plt.axhline(y=0.1, color='red', linestyle='--')
plt.text(3.55, 0.09, 'Baseline', fontsize=10, color='red')


# Set labels and title
plt.xlabel('Method', fontsize=16)
plt.ylabel('Hit@1', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.title('Model performance - Base dataset', fontsize=16)
plt.ylim(0, 1)
plt.tight_layout()
# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load df2 and df3
df = pd.read_csv('/home/antoine/gene_pheno_pred/figures/wandb_base_extended.csv')
df2 = pd.read_csv('/home/antoine/gene_pheno_pred/figures/wandb_go2.csv')
df3 = pd.read_csv('/home/antoine/gene_pheno_pred/figures/wandb_not-pheno.csv')
df4 = pd.read_csv('/home/antoine/gene_pheno_pred/figures/wandb_not_reified.csv')

# Combine all dataframes into a single dataframe
combined_df = pd.concat([df, df2, df3, df4], keys=['Base', 'GO', 'Not-Pheno', 'No reification'])

# Set up the plot
plt.figure(figsize=(8, 5))

# Set the bar width
bar_width = 0.2

# Create the bar plot for each method
method_order = ['TransE', 'TorusE', 'ComplEx', 'ConvKB']
for i, method in enumerate(method_order):
    method_data = combined_df[combined_df['method'] == method]
    plt.bar(
        x=i - bar_width * 1.5,
        height=method_data.xs('Base')['Hit@1'],
        width=bar_width,
        color='blue',
        label='Base'
    )
    plt.bar(
        x=i - bar_width * 0.5,
        height=method_data.xs('GO')['Hit@1'],
        width=bar_width,
        color='green',
        label='GO'
    )
    plt.bar(
        x=i + bar_width * 0.5,
        height=method_data.xs('Not-Pheno')['Hit@1'],
        width=bar_width,
        color='orange',
        label='Not-Pheno'
    )
    plt.bar(
        x=i + bar_width * 1.5,
        height=method_data.xs('No reification')['Hit@1'],
        width=bar_width,
        color='red',
        label='No reification'
    )

# Set labels and title
plt.xlabel('Method', fontsize=16)
plt.ylabel('Hit@1', fontsize=16)
plt.xticks(range(len(method_order)), method_order, fontsize=14)
plt.tick_params(axis='y', which='major', labelsize=14)
plt.title('Embedding quality for different data types', fontsize=18)

# Create custom legend
custom_legend = [
    plt.Line2D([0], [0], marker='s', color='blue', label='Base', linestyle=''),
    plt.Line2D([0], [0], marker='s', color='green', label='GO', linestyle=''),
    plt.Line2D([0], [0], marker='s', color='orange', label='Not-Pheno', linestyle=''),
    plt.Line2D([0], [0], marker='s', color='red', label='No reification', linestyle='')
]

# Add legend to the plot
plt.legend(handles=custom_legend, fontsize=14)

plt.ylim(0, 1)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load df2 and df3
df = pd.read_csv('/home/antoine/gene_pheno_pred/figures/wandb_base_extended.csv')
df2 = pd.read_csv('/home/antoine/gene_pheno_pred/figures/wandb_go2.csv')
df3 = pd.read_csv('/home/antoine/gene_pheno_pred/figures/wandb_not-pheno.csv')
df4 = pd.read_csv('/home/antoine/gene_pheno_pred/figures/wandb_not_reified.csv')

# Combine all dataframes into a single dataframe
combined_df = pd.concat([df, df2, df3, df4], keys=['Base', 'GO', 'Not-Pheno', 'No reification'])

# Set up the plot
plt.figure(figsize=(8, 5))

# Set the bar width
bar_width = 0.2

# Set the error bar width
error_bar_width = 0.02

# # Define a custom color palette
# custom_palette = ['purple', 'yellow', 'teal', 'red']

# # Set the color palette
# sns.set_palette(custom_palette)


# Create the bar plot for each method
method_order = ['TransE', 'TorusE', 'ComplEx', 'ConvKB']
for i, method in enumerate(method_order):
    method_data = combined_df[combined_df['method'] == method]
    base_data = method_data.xs('Base')['Hit@1']
    go_data = method_data.xs('GO')['Hit@1']
    not_pheno_data = method_data.xs('Not-Pheno')['Hit@1']
    no_reification_data = method_data.xs('No reification')['Hit@1']
    
    # Calculate the mean and standard deviation for each method
    mean_values = [base_data.mean(), go_data.mean(), not_pheno_data.mean(), no_reification_data.mean()]
    std_values = [base_data.std(), go_data.std(), not_pheno_data.std(), no_reification_data.std()]
    
    # Plot the bars with error bars
    sns.bar(
        x=i - bar_width * 1.5,
        height=mean_values[0],
        width=bar_width,
        label='Base'
    )
    plt.bar(
        x=i - bar_width * 0.5,
        height=mean_values[1],
        width=bar_width,
        label='GO'
    )
    plt.bar(
        x=i + bar_width * 0.5,
        height=mean_values[2],
        width=bar_width,
        label='Not-Pheno'
    )
    plt.bar(
        x=i + bar_width * 1.5,
        height=mean_values[3],
        width=bar_width,
        label='No reification'
    )
    
    # Add error bars
    plt.errorbar(
        x=i - bar_width * 1.5,
        y=mean_values[0],
        yerr=std_values[0],
        fmt='none',
        color='black',
        capsize=5
    )
    plt.errorbar(
        x=i - bar_width * 0.5,
        y=mean_values[1],
        yerr=std_values[1],
        fmt='none',
        color='black',
        capsize=5
    )
    plt.errorbar(
        x=i + bar_width * 0.5,
        y=mean_values[2],
        yerr=std_values[2],
        fmt='none',
        color='black',
        capsize=5
    )
    plt.errorbar(
        x=i + bar_width * 1.5,
        y=mean_values[3],
        yerr=std_values[3],
        fmt='none',
        color='black',
        capsize=5
    )

# Set labels and title
plt.xlabel('Method', fontsize=16)
plt.ylabel('Hit@1', fontsize=16)
plt.xticks(range(len(method_order)), method_order, fontsize=14)
plt.tick_params(axis='y', which='major', labelsize=14)
plt.title('Embedding quality for different data types', fontsize=18)

# Create custom legend
custom_legend = [
    plt.Line2D([0], [0], marker='s', color='blue', label='Base', linestyle=''),
    plt.Line2D([0], [0], marker='s', color='green', label='GO', linestyle=''),
    plt.Line2D([0], [0], marker='s', color='orange', label='Not-Pheno', linestyle=''),
    plt.Line2D([0], [0], marker='s', color='red', label='No reification', linestyle='')
]

# Add legend to the plot
plt.legend(handles=custom_legend, fontsize=14)

plt.ylim(0, 1)
plt.tight_layout()

# Show the plot
plt.show()

```
